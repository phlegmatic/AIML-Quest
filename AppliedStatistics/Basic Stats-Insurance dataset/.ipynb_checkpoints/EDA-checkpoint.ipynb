## Hypotheis Testing

### Do charges of people who smoke differ significantly from the people who don't?
Common sense says that smokers will be charged higher than than non-smokers lets try to prove our hypothesis statistically.
Let's box plot smoker vs charges to see how the charges distribution is in our dataset. 
T-test should be used for testing the significance.

```python
sns.boxplot(x="smoker",y="charges",data=rawInsuranceDf)
print("Box plot shows that Median of smokers is higher non-smokers")

#creating 
smokersDf=rawInsuranceDf[rawInsuranceDf['smoker']==1]
nonSmokersDf=rawInsuranceDf[rawInsuranceDf['smoker']==0]
smokersChargesDf=(smokersDf['charges'])
nonSmokersChargesDf=(nonSmokersDf['charges'])

#these are independent columns meaning a person cannot be smoker as well as non-smoker he must belong to one of the categories
from scipy import stats
z_statistic,p_value=stats.ttest_ind(smokersChargesDf,nonSmokersChargesDf)
print("zstatistic , p value of mannwhitneyu test is")
print(z_statistic,p_value)

def significance_test(num):
	"""This function is for comparing calculated p value agains 0.05 and retruning significance """
	if num <= 0.05:
		print(color.BOLD +"significantly different"+ color.END)
	else:
		print(color.BOLD +"not significantly different"+ color.END)
print("charges for people who smoke are ")
significance_test(p_value)

```

### Does bmi of males differ significantly from that of females?
We will repeat our above method for answering this question. 
Let's violin plot sex vs bmi to see how the bmi distribution is in our dataset. 

```python
from scipy import stats

sns.violinplot(x="sex",y="bmi",data=rawInsuranceDf)

maleDf=rawInsuranceDf[rawInsuranceDf['sex']==0]
femaleDf=rawInsuranceDf[rawInsuranceDf['sex']==1]
maleBMIDf=(maleDf['bmi'])
femalBMIeDf=(femaleDf['bmi'])
z_statistic,p_value=stats.ttest_ind(maleBMIDf,femalBMIeDf)
print("zstatistic , p value of mannwhitneyu test is")
#z_statistic,p_value=mannwhitneyu(maleBMIDf,femalBMIeDf)
print(z_statistic,p_value)

print("bmi of males vs females is")
significance_test(z_statistic)
```

```python
sns.heatmap(rawInsuranceDf.corr(), annot=True)
print("Below Correlation heatmap is also supporting above two tests as \n"
      +"correlation between smoker and charges is high (0.79) \n"
      +"correlation between sex and bmi is low (-0.046)")
```

### Is the proportion of smokers significantly different in different genders?
Let's observe the countplot to see the actual counts. we can use chi-square test against crosstabulated data indicating proportions.

```python
sns.catplot(x="sex", hue="smoker", kind="count", data=rawInsuranceDf);

#As we are trying two compare two categorical variable ,chi-square is best suited. 
pd.crosstab(rawInsuranceDf['sex'],rawInsuranceDf['smoker'] )
chiAnalysisDf = np.array([[517,159],[547,115]]) #crosstab rows
chi_sq_Stat, p_value, deg_freedom, exp_freq = scipy.stats.chi2_contingency(chiAnalysisDf)

print('Chi-square statistic %3.5f P value %1.6f Degrees of freedom %d' %(chi_sq_Stat, p_value,deg_freedom))
print("Proportion of smokers across genders is")
significance_test(p_value)
```

### Is the distribution of bmi across women with no children, one child and two children, the same ?
Let's boxplot the distributions and compare the median and spread of the distribution. We will use Kolmogorov-Smirnov statistic on 2 samples to compare the distributions. https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.ks_2samp.html
https://www.listendata.com/2019/07/KS-Statistics-Python.html

```python
plt.title("Box plot for bmi of women based on number of children")
sns.boxplot(y="children", x="bmi", data= rawInsuranceDf[(rawInsuranceDf.sex==1)],orient="h",palette ='magma')

childrenBmiFemaleDf  = femaleDf[['bmi','children']]

d0 =childrenBmiFemaleDf.loc[childrenBmiFemaleDf.children==0][['bmi']]
d1 = childrenBmiFemaleDf.loc[childrenBmiFemaleDf.children==1][['bmi']]
d2 =  childrenBmiFemaleDf.loc[childrenBmiFemaleDf.children==2][['bmi']]

#preparing 1d array for passing to test
a0 = np.empty(d0.shape[0], dtype=object)
a1 = np.empty(d1.shape[0], dtype=object)
a2 = np.empty(d2.shape[0], dtype=object)

a0[:] = d0.values.tolist()
a1[:] = d1.values.tolist()
a2[:] = d2.values.tolist()


from scipy.stats import ks_2samp
print("Ks 2 sample test outcome between distribution of bmi across women with no child, one child")
ks_2samp(a0,a1)
```

```python
print("Ks 2 sample test outcome between distribution of bmi across women with no child, two children")
ks_2samp(a0,a2)
```

```python
print("Ks 2 sample test outcome between distribution of bmi across women with two children, one child")
ks_2samp(a1,a2)
```

```python
print(" For an similar distribution, we cannot reject the null hypothesis since the p-value is high 92%,29%,37% respectively")
```
